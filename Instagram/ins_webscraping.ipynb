```python
import pandas as pd
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
client = ApifyClient("apify_api_0RMlhCC3PcIjYvp8g3Ga47hdjRVsKf3cSQNB")

# Prepare the Actor input
run_input = {
    "directUrls": [
        "https://www.instagram.com/p/CpNvdBKOrIO/?img_index=1",
    ],
    "resultsLimit": 24,
}

# Run the Actor and wait for it to finish
run = client.actor("apify/instagram-comment-scraper").call(run_input=run_input)
```

### Here I scrape all the comments and save to an excel file

```python
# Fetch and print Actor results from the run's dataset (if there are any)
Comments_ins = []
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
    Comment_ins = item['text']
    print(Comment_ins)
    Comments_ins.append(Comment_ins)
print(Comments_ins)
df2 = pd.DataFrame(Comments_ins, columns=['Comments'])
df2.to_excel('ins_comments.xlsx', index=False)
```
